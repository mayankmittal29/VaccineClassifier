### Now as 3271 is too much from 26707 rows so cant just drop these rows having atleast one nan so will try to impute values ###

```python
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.model_selection import train_test_split
```

```python
imputecols = ['xyz_concern','xyz_knowledge','doctor_recc_xyz','doctor_recc_seasonal','chronic_med_condition','health_worker','opinion_xyz_vacc_effective','opinion_xyz_risk','opinion_seas_vacc_effective','opinion_seas_risk']
imputer = SimpleImputer(strategy='mean')
finaldata[imputecols] = imputer.fit_transform(finaldata[imputecols])
# Features and targets
X = finaldata[imputecols]
y_xyz = finaldata['xyzprob']
y_seas = finaldata['seasprob']
```

### Now algorithms for binary classification are : Logistic Regression , K-Nearest Neighbours , SVM, Decision Trees and Random Forest ###

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Assuming finaldata is your dataframe and you have already handled missing values
imputecols = ['xyz_concern','xyz_knowledge','doctor_recc_xyz','doctor_recc_seasonal','chronic_med_condition','health_worker','opinion_xyz_vacc_effective','opinion_xyz_risk','opinion_seas_vacc_effective','opinion_seas_risk']

# Combine features and targets
X = finaldata[imputecols + ['xyzprob', 'seasprob']]

# Split the data into training and test sets without shuffling
train_data, test_data = train_test_split(X, test_size=0.2, random_state=42, shuffle=False)

# Separate features and targets for training set
X_train = train_data[imputecols]
y_train_xyz = train_data['xyzprob']
y_train_seas = train_data['seasprob']

# Separate features and targets for test set
X_test = test_data[imputecols]
y_test_xyz = test_data['xyzprob']
y_test_seas = test_data['seasprob']

# Initialize the models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(probability=True),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Naive Bayes": GaussianNB()
}

# Function to train, predict and calculate metrics
def train_predict_evaluate(models, X_train, y_train, X_test, y_test):
    results = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred_proba = model.predict_proba(X_test)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        results[name] = {
            "fpr": fpr,
            "tpr": tpr,
            "thresholds": thresholds,
            "roc_auc": roc_auc
        }
    return results

# Train and evaluate models for xyzprob
results_xyz = train_predict_evaluate(models, X_train, y_train_xyz, X_test, y_test_xyz)

# Train and evaluate models for seasprob
results_seas = train_predict_evaluate(models, X_train, y_train_seas, X_test, y_test_seas)

# Function to plot ROC curves
def plot_roc(results, title):
    plt.figure(figsize=(10, 8))
    for name, result in results.items():
        plt.plot(result["fpr"], result["tpr"], label=f'{name} (AUC = {result["roc_auc"]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()

# Plot ROC curves for xyzprob
plot_roc(results_xyz, 'ROC Curve for xyzprob')

# Plot ROC curves for seasprob
plot_roc(results_seas, 'ROC Curve for seasprob')
```

```python
from prettytable import PrettyTable
def display_pretty_table(results):
    table = PrettyTable()
    table.field_names = ["Model", "AUC", "TPR", "FPR", "Thresholds"]
    
    for name, metrics in results.items():
        fpr = metrics["fpr"]
        tpr = metrics["tpr"]
        roc_auc = metrics["roc_auc"]
        thresholds = metrics["thresholds"]
        
        # Find the threshold where TPR is closest to 0.8 (example for demonstration)
        closest_tpr_index = min(range(len(tpr)), key=lambda i: abs(tpr[i]-0.8))
        closest_threshold = thresholds[closest_tpr_index]
        
        table.add_row([name, f'{roc_auc:.4f}', f'{tpr[closest_tpr_index]:.4f}', f'{fpr[closest_tpr_index]:.4f}', f'{closest_threshold:.4f}'])
    
    print("ROC Curve Metrics:")
    print(table)


# Display results in tabular form using PrettyTable
print("Results for xyzprob:")
display_pretty_table(results_xyz)

print("\nResults for seasprob:")
display_pretty_table(results_seas)
```

### So on above analysis, We can say that for XYZ vaccine- Logistic Regression model is best and for seasonal vaccine- SVM model is best ###

```python
# Evaluate AUC scores on the new test data
testdataset = pd.read_csv('test_set_features.csv')
finaltestdata = testdataset[imputecols]
imputer = SimpleImputer(strategy='mean')
finaltestdata[imputecols] = imputer.fit_transform(finaltestdata[imputecols])
def evaluate_auc_on_new_data(models,new_test_data,flag):
    modelname =''
    if flag == 'XYZ':
       modelname = 'Logistic Regression'
    elif flag == 'SEAS':
        modelname ='SVM'
    for name,model in models.items():
       if name == modelname:
        return model.predict_proba(new_test_data)[:, 1]

best_model_xyz = evaluate_auc_on_new_data(models, finaltestdata,'XYZ')
best_model_seas = evaluate_auc_on_new_data(models,finaltestdata,'SEAS')
```

### Now storing results in new CSV file ###


```python
selected_data = pd.DataFrame({
    'respondent_id': testdataset['respondent_id'],
    'xyz_vaccine': best_model_xyz,
    'seasonal_vaccine': best_model_seas
})

# Save to CSV
output_file = 'predictions_vacc.csv'
selected_data.to_csv(output_file, index=False)
selected_data.head(10)
```
